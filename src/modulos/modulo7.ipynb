Veamos c贸mo puedes crear c贸digo para pedir a los usuarios que escriban valores y, despu茅s, permitirles usar `done` cuando terminen de escribir los valores. En nuestro ejemplo, usaremos la entrada de usuario como condici贸n y, despu茅s, la probaremos al principio del ciclo `while`.

```python
user_input = ''

while user_input.lower() != 'done':
    user_input = input('Enter a new value, or done when done')
```

```python
# Creamos la variable que almacena el texto
user_input = ''
# Creamos la lista que almacena cada uno de los textos que el usuario ingresa
inputs = []

# Ciclo while
while user_input.lower() != 'done':
    # Verificamos si hay un valor en user_input
    if user_input:
        # Almacenamos ese valor en la lista
        inputs.append(user_input)
    # Capturamos un nuevo valor
    user_input = input('Enter a new value, or done when done')
```

## Uso de Ciclos `for` con listas

```python
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]

print("The first planet is ", planets[0])
print("The second planet is ", planets[1])
print("The third planet is ", planets[2])

countdown = [4, 3, 2, 1, 0]
for number in countdown:
    print(number)
print("Blast off!! ")
```

Vamos a cambiar el c贸digo anterior para que espere un segundo entre cada n煤mero mediante la funci贸n `sleep()`:

```python
# De la biblioteca time, importamos (traemos) la clase sleep

from time import sleep

# Creamos una lista de 5 n煤meros llamada countdown
countdown = [4, 3, 2, 1, 0]

# Para cada n煤mero en countdown
for number in countdown:
    #Muestra el n煤mero
    print(number)

    # Espera (1segundo)
    sleep(1)

# Muestra el mensaje Blast off
print("Blast off!! ")
```
